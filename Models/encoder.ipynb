```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
db  = pd.read_csv('db_mod.csv')
import pandas as pd
import re
import string
from bs4 import BeautifulSoup
import nltk
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def remove_punctuation(text):
    # Iterate through each character in the text
    no_punct = [char for char in text if char not in string.punctuation]
    # Join the characters back into a string
    words_wo_punct = ''.join(no_punct)
    return words_wo_punct

db['statement'] = db['statement'].apply(lambda x: remove_punctuation(x))
db.head()
```

```python
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()  # Split the sentence into words
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)  # Join the filtered words back into a sentence

db['statement'] = db['statement'].apply(lambda x: remove_stopwords(x))
db.head()
```

```python
def to_lowercase(text):
    return text.lower()  # Convert the entire sentence to lowercase
db['statement'] = db['statement'].apply(lambda x: to_lowercase(x))  # Apply lowercase conversion
db.head()
```

```python
mlb = MultiLabelBinarizer()
```

```python
# Assuming your tags are in a column named 'tags' and are comma-separated
tags_data = db['tags'].str.split(',')
encoded_tags = mlb.fit_transform(tags_data)
```
