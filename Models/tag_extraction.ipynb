```python
import numpy as np
import pandas as pd
```

```python
db = pd.read_csv('problems.csv')
```

```python
db.head(10)
```

```python
db.info()
```

```python
db = db.drop(['url', 'contestId', 'index' , 'result'], axis=1)
```

```python
for i in range(len(db)):
    # Use .loc to access rows by their index and update the 'uid' column
    db.loc[i, 'uid'] = db.loc[i, 'uid'].replace('Codeforces', '')
```

```python
db.drop_duplicates(subset=['uid'], keep='first', inplace=True)
```

```python
db['tags'] = db['tags'].astype(str)

target_tags = ['dp', 'bitmasks', 'greedy','probabilities', 'games', 'interactive' ,'graphs',
               'combinatorics', 'flows' , 'sortings']  # Replace with your desired tags
tag_filter = db['tags'].apply(lambda x: any(tag in x for tag in target_tags))
```

```python
db_mod = db[tag_filter]
```

```python
db_mod.info()
```

```python
db_mod = db_mod.astype(str)
```

```python
# Filter the DataFrame to get the row with uid '1895G'
row_1895G = db_mod[db_mod['uid'] == '1895G']

# Display the original 'tags' value for this row
print(row_1895G['tags'].iloc[0])
```

```python
import numpy as np

def filter_tags(tags_str):
    if isinstance(tags_str, str):
        tags_str = tags_str.lower()
        tags_list = [tag.strip() for tag in tags_str.split(',')] # Strip spaces here
        filtered_tags = [tag for tag in tags_list if tag in [t.strip() for t in target_tags]] # Strip spaces here
        if filtered_tags:
            return ','.join(filtered_tags)
        else:
            return np.nan
    else:
        return np.nan

db_mod['filtered_tags'] = db_mod['tags'].apply(filter_tags)
db_mod['tags'] = db_mod['filtered_tags']
db_mod = db_mod.drop('filtered_tags', axis=1)
```

```python
# Filter the DataFrame to get the row with uid '1895G'
row_1895G = db_mod[db_mod['uid'] == '1895G']

# Display the original 'tags' value for this row
print(row_1895G['tags'].iloc[0])
```
